```python
import pandas as pd
 
df=pd.read_csv('diabetes.csv')
 
df.info()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=6)
```

```python
x= df.iloc[:,1:4]
print(x)
```

```python
fig =plt.figure()
ax= fig.add_subplot(projection='3d')
ax.scatter(x.Glucose,x.BloodPressure,x.SkinThickness)
ax.set_xlabel('Glucose')
ax.set_ylabel('BloodPressure')
ax.set_zlabel('SkinThickness')
```

```python
kmeans.fit(x)
```

```python
import pandas as pd
df=pd.read_csv('diabetes.csv')
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=3)
```

```python
x= df.iloc[:,1:4]
print(x)
```

```python
ac.fit(x)
```

```python
fig =plt.figure()
ax= fig.add_subplot(projection='3d')
ax.scatter(x.Glucose,x.BloodPressure,x.SkinThickness,c=ac.labels_,cmap='rainbow')
ax.set_xlabel('Glucose')
ax.set_ylabel('BloodPressure')
ax.set_zlabel('SkinThickness')
```

```python
ac = AgglomerativeClustering(n_clusters=4)
```

```python
x= df.iloc[:,1:4]
print(x)
```

```python
ac.fit(x)
```

```python
fig =plt.figure()
ax= fig.add_subplot(projection='3d')
ax.scatter(x.Glucose,x.BloodPressure,x.SkinThickness,c=ac.labels_,cmap='rainbow')
ax.set_xlabel('Glucose')
ax.set_ylabel('BloodPressure')
ax.set_zlabel('SkinThickness')
```
